```python
import pandas as pd
import http.client
import urllib.parse as urlparse
import sys
sys.path.insert(1, 'User Profile Overview')
import APPLICATION_ID, REST_API_KEY, READ_ONLY_MASTER_KEY
import json
connection = http.client.HTTPSConnection('jooay.herokuapp.com', 443)
connection.connect()
connection.request('GET', '/parse/classes/_User?'+urlparse.urlencode({"limit": 10000}), '', {
    "X-Parse-Application-Id": APPLICATION_ID,
    "X-Parse-REST-API-Key": REST_API_KEY,
    "X-Parse-Master-Key": READ_ONLY_MASTER_KEY,
})

users = json.loads(connection.getresponse().read())
users = pd.DataFrame(users['results'])
users[users['newsletter']==True].to_excel("newsletter_signees.xlsx")
```
